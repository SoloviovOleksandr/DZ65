```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
from imutils import face_utils
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
face_cascade = cv2.CascadeClassifier('/content/gdrive/MyDrive/DZ65/haarcascade_frontalface_default.xml')
img = cv2.imread('/content/gdrive/MyDrive/DZ65/titanic.jpg')
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis("off")
plt.show;
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x, y, w, h) in faces:
    cv2.rectangle(img1, (x, y), (x+w, y+h), (255, 0, 0), 5)
plt.imshow(img1)
plt.axis('off')
plt.show;
```

```python
img90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
gray = cv2.cvtColor(img90, cv2.COLOR_BGR2GRAY)
img11 = cv2.cvtColor(img90, cv2.COLOR_BGR2RGB)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x, y, w, h) in faces:
    cv2.rectangle(img11, (x, y), (x+w, y+h), (255, 0, 0), 5)
plt.imshow(img11)
plt.axis('off')
plt.show;
```

```python
face_detect = dlib.get_frontal_face_detector()
rects = face_detect(img1, 1)

for (i, rect) in enumerate(rects):
    (x, y, w, h) = face_utils.rect_to_bb(rect)
    cv2.rectangle(img1, (x, y), (x + w, y + h), (255, 255, 255), 3)
    
plt.figure(figsize=(10,6))
plt.imshow(img1, cmap='gray')
plt.axis('off')
plt.show();
```
